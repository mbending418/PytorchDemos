```python
from pytorch_demos.NMIST.simple_fcn import FCNetImageClassifier
from pytorch_demos.NMIST.simple_fcn_trainer import train_fcn
from pytorch_demos.NMIST.simple_cnn import ConvNet
from pytorch_demos.NMIST.simple_cnn_trainer import train_cnn

from torchvision import datasets, transforms

import os
import torch

saved_model_directory = os.path.join("..","saved_models")
saved_fcn_model_name = "fcn_nmist_demo"
saved_cnn_model_name = "cnn_nmist_demo"
```

```python
#train fcn model

train_fcn(save_model=True,
          output_directory=saved_model_directory,
          output_filename=saved_fcn_model_name,
          log_interval=180)
```

```python
#reload the trained model
model = FCNetImageClassifier.load_model(file_name=saved_fcn_model_name,
                                        folder=saved_model_directory)

#set up data
image_mean = 0.13066047627384256  # mean value of the images in our training data
image_std =  0.3050502568821657  # standard deviation of the images in our training data

transform = transforms.Compose(
        [transforms.ToTensor(), transforms.Normalize((image_mean,), (image_std,))]
)
test_data = datasets.MNIST(
        "../data", train=False, download=True, transform=transform
)

raw_test_data = datasets.MNIST(
        "../data", train=False, download=True)

index = 15 #test the model on test data #15
datum = test_data[index][0].unsqueeze(1)
expected_result = test_data[index][1]
raw_test_datum = raw_test_data[index][0]

loader = torch.utils.data.DataLoader(datum, batch_size=1)

model_output = int(model(datum).argmax(dim=1, keepdim=True))

print(f"model_output: {model_output}")
print(f"expected_result: {expected_result}")
raw_test_datum
```

```python
#train cnn model

train_cnn(save_model=True,
          output_directory=saved_model_directory,
          output_filename=saved_cnn_model_name,
          log_interval=180)
```
